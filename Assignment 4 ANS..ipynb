```python
class Computation:
    def __init__(self):
        pass
    def factorial(self, n):
        if n < 0:
            return None  
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result
    def sum_of_n(self, n):
        return n * (n + 1) // 2
    def test_prime(self, num):
        if num <= 1:
            return False
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                return False
        return True
    def test_primes(self, num1, num2):
        return self.test_prime(num1) and self.test_prime(num2)
    def table_multiplication(self, n):
        for i in range(1, 11):
            print(f"{n} x {i} = {n * i}")
    def all_tables_multiplication(self):
        for i in range(1, 10):
            print(f"\nMultiplication Table of {i}:")
            self.table_multiplication(i)
    @staticmethod
    def list_divisors(n):
        Ldiv = []
        for i in range(1, n + 1):
            if n % i == 0:
                Ldiv.append(i)
        return Ldiv
    @staticmethod
    def list_prime_divisors(n):
        Ldiv = []
        for i in range(2, n + 1):
            if n % i == 0 and all(i % j != 0 for j in range(2, int(i**0.5) + 1)):
                Ldiv.append(i)
        return Ldiv
comp = Computation()
print("Factorial of 5:", comp.factorial(5))
print("Sum of first 5 integers:", comp.sum_of_n(5))
print("Is 17 prime?", comp.test_prime(17))
print("Are 17 and 23 prime between them?", comp.test_primes(17, 23))
print("\nMultiplication Table of 7:")
comp.table_multiplication(7)
print("\nAll Multiplication Tables:")
comp.all_tables_multiplication()
print("\nDivisors of 12:", comp.list_divisors(12))
print("Prime divisors of 12:", comp.list_prime_divisors(12))
```

```python
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    def area(self):
        return self.width * self.height
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return math.pi * self.radius**2
class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height
    def area(self):
        return 0.5 * self.base * self.height
def calculate_total_area(shapes):
    total_area = 0
    for shape in shapes:
        area = shape.area()
        total_area += area
        print(f"Area of {type(shape).__name__}: {area}")
    return total_area
rectangle = Rectangle(5, 4)
circle = Circle(3)
triangle = Triangle(6, 2)
shapes = [rectangle, circle, triangle]
total_area = calculate_total_area(shapes)
print("Total area of shapes:", total_area)
```
